```python
# i.1 — Tabular In-Depth (presets, bagging/stacking, refit_full) — ONE CELL
!pip -q install -U pip setuptools wheel
!pip -q install -U autogluon

from autogluon.tabular import TabularDataset, TabularPredictor
train_url = 'https://autogluon.s3.amazonaws.com/datasets/Inc/train.csv'
test_url  = 'https://autogluon.s3.amazonaws.com/datasets/Inc/test.csv'
train = TabularDataset(train_url)
test  = TabularDataset(test_url)
label = 'class'

predictor = TabularPredictor(label=label).fit(
    train_data=train,
    presets="medium_quality_faster_train",
    num_bag_folds=5,
    num_stack_levels=1,
    refit_full=True,
    time_limit=600
)
try:
    from IPython.display import display
    display(predictor.leaderboard(test, silent=True))
except Exception:
    print(predictor.leaderboard(test, silent=True))
```

```python
# Multimodal Tabular (LITE: text + categorical + numeric, CPU-only, no downloads)
import numpy as np, pandas as pd, random
from autogluon.tabular import TabularPredictor
from sklearn.model_selection import train_test_split

# ---- Synthetic Pet-like data (small & fast) ----
rng = np.random.default_rng(0); random.seed(0)
N = 1200
breeds = ["mix","labrador","siamese","bulldog","persian"]
colors = ["black","white","brown","gold","tabby","tri"]
tokens_pos = ["friendly","playful","gentle","trained","housebroken","affectionate"]
tokens_neg = ["timid","special_needs","anxious","senior","requires_experience","shy"]

age = rng.integers(1, 180, size=N)                   # months
fee = np.round(rng.uniform(0, 300, size=N), 2)
breed = rng.choice(breeds, size=N)
color = rng.choice(colors, size=N)
photo_amt = rng.integers(0, 6, size=N)

def make_desc():
    words = rng.choice(tokens_pos, size=rng.integers(2,4)).tolist()
    if rng.random() < 0.4:
        words += [rng.choice(tokens_neg)]
    return " ".join(words)

desc = [make_desc() for _ in range(N)]

has_trained = np.array(['trained' in s for s in desc], dtype=float)
has_senior  = np.array(['senior'  in s for s in desc], dtype=float)

# Score -> 5 adoption speed classes (0 fast .. 4 slow)
score = (-0.02*age) + (0.005*fee) - (0.15*(photo_amt>=3).astype(float)) - 0.3*has_trained + 0.2*has_senior
bins = np.quantile(score, [0, .2, .4, .6, .8, 1.0])
y = np.digitize(score, bins[1:-1], right=True)  # 0..4

df = pd.DataFrame({
    "Age": age,
    "Fee": fee,
    "Breed": breed,
    "Color": color,
    "Description": desc,      # <-- TEXT modality
    "PhotoAmt": photo_amt,
    "AdoptionSpeed": y.astype(int)
})

train_df, test_df = train_test_split(df, test_size=0.2, random_state=42, stratify=df["AdoptionSpeed"])

# ---- Train tiny AutoGluon model (fast presets + short time_limit) ----
predictor = TabularPredictor(
    label="AdoptionSpeed",
    problem_type="multiclass",
    eval_metric="accuracy"
).fit(
    train_data=train_df,
    presets="medium_quality_faster_train",
    time_limit=90  # keep it snappy
)

print("Holdout accuracy:", predictor.evaluate(test_df))
try:
    from IPython.display import display
    display(predictor.leaderboard(test_df, silent=True).head(10))
except Exception:
    print(predictor.leaderboard(test_df, silent=True).head(10))
```

```python
# i.3 — Automatic Feature Engineering (AutoMLPipelineFeatureGenerator) — ONE CELL
!pip -q install -U pip setuptools wheel
!pip -q install -U autogluon scikit-learn

import numpy as np, random, pandas as pd
from datetime import datetime
from sklearn.datasets import make_regression
from autogluon.tabular import TabularPredictor
from autogluon.features.generators import AutoMLPipelineFeatureGenerator

# Synthetic mixed-type data
X, y = make_regression(n_samples=600, n_features=5, random_state=1, noise=10.0)
dfx = pd.DataFrame(X, columns=list("ABCDE"))
dfx['B'] = (dfx['B']*10).astype(int)
dfx['C'] = pd.to_datetime(datetime(2000,1,1)) + pd.to_timedelta((dfx['C']*20).astype(int), unit='D')
bins = [-np.inf,-5,0,5,np.inf]; labels = ['v','w','x','y']
dfx['D'] = pd.cut(dfx['D']*5, bins, labels=labels)
vocab = ["alpha","beta","gamma","delta","epsilon","zeta"]
random.seed(0)
dfx['E'] = [" ".join(random.choice(vocab) for _ in range(5)) for __ in range(len(dfx))]
df = dfx.copy(); df['label'] = y

predictor = TabularPredictor(label='label', problem_type='regression').fit(
    train_data=df,
    feature_generator=AutoMLPipelineFeatureGenerator(),
    time_limit=180
)
fi = predictor.feature_importance(df)
try:
    from IPython.display import display
    display(fi.head(15))
except Exception:
    print(fi.head(15))
```
